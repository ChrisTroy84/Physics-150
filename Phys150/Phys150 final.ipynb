```python
x = input("Enter a string:")
print(x[::-1])     
```

```python
def pal(x):
    if x[::-1] == x:
        return True
    else:
        return False
pal('dad')
```

5. Write a function that takes an argument $n$ and outputs a list containing all of its factors. (4pt)

 - example: with argument n=10, the function should output [1,2,5,10]

```python
def factors(n):
    num=100
    nlist=[]
    for x in range(1,num+1):
        if n%x==0:
            nlist.append(x)
    print(nlist)

factors(10)
```

6. Write a function to calculate the triple factorial of a number. The triple factorial of n, is the product of positive integers less than or equal to n and congruent to n mod 3. For example, the triple factorial of 8 is $8\times 5\times 2=80$ and the triple factorial of 9 is $9\times 6\times 3\times 1=162$. Hint: You may solve this problem use a while loop and think about the condition for the loop to terminate. (4pt)

```python
def trip_factorial(n):
    if n == 0 or n == 1:
         return 1
    else:
         return n * (n-3) * (n-6)
trip_factorial(8)
```

7. A ball is launched upwards with an initial velocity of $10m/s$. Due to gravity, the acceleration of the ball is $-10 m/s^2$ and remains constant. We take the initial position of the ball to be zero. Simulate the motion of the ball before it falls back to the initial position using the Euler method. Plot the velocity $v$ and position $h$ as functions of time on the same plot. Adding labels and legends. (4pt)

```python
import matplotlib.pyplot as plt
import math

def motion(t_max, num_step, t=0, x=0, vx=0, vy=0, a=0):
    time_list=[t]
    pos_list=[x]
    dt=t_max/num_step
    for i_step in range(num_step):
        t = t + dt
        vx = vx 
        vy = vy + a*dt
        v = math.sqrt(vx**2 + vy**2)
        x = x + vy*dt
        if x < 0:
            v = -v
            break
        time_list.append(t)
        pos_list.append(x)
    return time_list, pos_list,
```

```python
#x velocity = 6, y-velocity = 8, so v = 10 m/s
time_list, pos_list,= motion(t_max=10,num_step=1000,t=0,x=0,vx=6,vy=8,a=-10)
plt.plot(time_list,pos_list)
plt.xlabel('time')
plt.ylabel('position')
plt.legend(('position','velocity'))
```

8. Simulate the projectile motion of a ball. The ball is launched with a horizontal velocity $5m/s$ and vertical velocity $10m/s$. Take the gravity accelaration to be $-10m/s^2$. The ball bounces back when it hits the ground. 
Simulate the motion for duration of 5 seconds. Plot the trajectory and make an animation of the motion. (5pt)

```python
import matplotlib.pyplot as plt
import math

def motion(t_max, num_step, t=0, x=0, vx=0, vy=0, a=0):
    time_list=[t]
    pos_list=[x]
    dt=t_max/num_step
    for i_step in range(num_step):
        t = t + dt
        vx = vx 
        vy = vy + a*dt
        v = math.sqrt(vx**2 + vy**2)
        x = x + vy*dt
        if x < 0:
            vx = vx 
            vy = vy + a*dt
            v = math.sqrt(vx**2 + vy**2)
            x = x + vy*dt
            v = -v
            break
        time_list.append(t)
        pos_list.append(x)
    return time_list, pos_list
```

```python
time_list, pos_list = motion(t_max=5,num_step=1000,t=0,x=0,vx=5,vy=10,a=-10)
plt.plot(time_list,pos_list)
plt.xlabel('time')
plt.ylabel('position')
```

```python
plt.xlabel('x')
plt.ylabel('y')

fig = plt.gcf()
ax = plt.gca()
ax.axis('equal')

circle1 = plt.Circle((0,0), radius=0.002, color='tab:blue')
ax.add_patch(circle1)

def draw_frame(frame_indx):
    lines = ax.get_lines()
    if frame_indx < len(time_list):
        lines[0].set_data(time_list[0:frame_indx],pos_list[0:frame_indx])
        circle1.center = (time_list[frame_indx], pos_list[frame_indx])
```

```python
import IPython
import time
for frame_indx in range(1000000):
    draw_frame(frame_indx)
    IPython.display.display(fig)
    IPython.display.clear_output(wait=True)
    time.sleep(0.01) 

draw_frame(1000000)
```
