2. Input a integer number using input() function. Check if the number is even or odd and print a message.  (2pt)

```python
x = input("Please enter an integer value: ")

x = int(x)

if (x % 2 == 0):
    print("Your number is even")
elif ((x % 2) >= 1):
    print("Your number is odd")
else:
    print("Invalid input, please try again")
```

3. Make a program converting miles to kilometers and vice versa. Use __one__ input() function for input. When the user input is distance in miles (a number followed by mi), convert it to kilometers. When the user inputs is distance in kilemeters (a number followed by km), convert it to miles. Print the result to the screen. If the user input neither ends with 'mi' or 'km', print "incorrect format". (Hint: Recall that you can get the last two characters of a string a by a[-2:] and everything before the last two characters by a[:-2] ) (5pt)

_example_ :

input: 1 mi

output: 1.60 km

input: 1 km
output: 0.62 mi

input: 1 m
output: incorrect format

```python
x = input("Enter an amount of distance ending in mi (miles) or km (kilometers): ")

if x[-2:] == "mi":
    x = int(x[:-2])
    print("That is", (x * 1.60934), "kilometers")
elif x[-2:] == "km":
    x = int(x[:-2])
    print("That is", (x * 0.62137), "miles")
elif x[-2:] != "mi" or "km":
    print("Incorrect format")
```

4. Check password: Input a password from the keyboard using input() function. Check if the password meets the following requirements. (8pt)
- at least 8 characters, at most 20 characters
- at least one uppercase letter (A-Z)  
- at least one number (0-9)
- at least one special character from ! @ # $ % ^ & * ()

Print "valid password" if the password meets all the requirements and print "invalid password" otherwise

```python
num="1234567890"
alpha="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
specialchar = '!@#$%^&*()'

num = set(num)
alpha = set(alpha)
specialchar = set(specialchar)

password = input("Please enter your password: ")
password = set(password)

if len(password) < 8 or len(password) > 20:
    print("invalid password")
elif password.isdisjoint(num) or password.isdisjoint(alpha) or password.isdisjoint(specialchar):
    print("invalid password")
else:
    print("valid password")
```

5. Come up with an interesting coding problem and solve it using branching. Be creative! (4pt)

Make a program that calculates the final grade of a student. Take 3 inputs: exams, homework, and final project.
Print the students letter grade based on the requirements.

Requirements:

Grade Scale
- A: 90-100
- B: 80-89
- C: 70-79
- F: <=69

Percentages:
- Exams = 60%
- homework = 20%
- final project = 20%

```python
x = input("Enter the student's exam average: ")
y = input("Enter the student's homework score: ")
z = input("Enter the student's final project score: ")

x = float(x)
y = float(y)
z = float(z)

grade = ((x*0.6) + (y*0.2) + (z*0.2))
grade = float(grade)
    
if grade >= 90:
    print("This student receives an A")
elif grade >= 80 and grade <= 89:
    print("This student receives an B")
elif grade >= 70 and grade <= 79:
    print("This student receives an C")
elif grade <= 69:
    print("This student receives an F")
```
