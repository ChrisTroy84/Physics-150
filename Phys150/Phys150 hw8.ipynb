1. Check password with a function: write a function to check if the password meets the following requirements. (6pt)
    - at least 8 characters, at most 20 characters
    - at least one uppercase letter (A-Z)  
    - at least one number (0-9)
    - at least one special character from ! @ # $ % ^ & * ()

    __The function should take a password as an argument and return true or false.__


```python
def password_check(password):
    if len(password)<8 or len(password)>20:
        return False
    elif set(password).isdisjoint(set('ABCDEFGHIJKLMNOPQRSTUVWXYZ')):
        return False
    elif set(password).isdisjoint(set('0123456789')):
        return False
    elif set(password).isdisjoint(set('!@#&%^&*()')):
        return False
    return True
    
password_check('ThisisPhys150!')
```

3. Simulate the projectile motion of a ball with initial horizontal velocity $v_x = 3 m/s$ and vertical velocity $v_y = 4 m/s$, the gravitational constant is $g=-10 m/s^2$. Plot the trajectory before the ball hitting the ground. (6pt)

```python
import matplotlib.pyplot as plt
import math
```

```python
def motion(t_max, num_step, t=0, x=0, vx=0, vy=0, a=0):
    time_list=[t]
    pos_list=[x]
    dt=t_max/num_step
    for i_step in range(num_step):
        t = t + dt
        vx = vx 
        vy = vy + a*dt
        v = math.sqrt(vx**2 + vy**2)
        x = x + vy*dt
        if x < 0:
            break
        time_list.append(t)
        pos_list.append(x)
    return time_list, pos_list
```

```python
time_list, pos_list = motion(t_max=1,num_step=100,t=0,x=0,vx=3,vy=4,a=-10)
plt.plot(time_list,pos_list)
plt.xlabel('time')
plt.ylabel('position')
```

You can import cos() and sin() function from the math module or the numpy module.
Try different angles see which leads to furthest travel distance.

```python
import matplotlib.pyplot as plt
import math
```

```python
def motion(t_max, num_step, t=0, x=0, v=0, z=0, a=0):
    time_list=[t]
    pos_list=[x]
    dt=t_max/num_step
    for i_step in range(num_step):
        t = t + dt
        vx = v*math.cos(z) 
        vy = (v*math.sin(z)) + a*dt
        v = math.sqrt((vx**2) + (vy**2))
        x = x + v*dt
        if x < 0:
            break
        time_list.append(t)
        pos_list.append(x)
    return time_list, pos_list
#z = angle
```

```python
time_list, pos_list = motion(t_max=1,num_step=100,t=0,x=0,v=5,z=45,a=-10)
plt.plot(time_list,pos_list)
plt.xlabel('time')
plt.ylabel('position')
```
